```python
import pandas as pd
import numpy as np
import random
import os

# =====================================================
# 1. PEMANGGILAN DATA ASLI
# =====================================================
df_real = pd.read_csv("scrapped_data_honkai_star_rail.csv")

print("üìÑ Nama kolom dataset asli:")
print(df_real.columns.tolist())
print(f"Jumlah total data asli: {len(df_real)}")

# =====================================================
# 2. DETEKSI KOLOM ULASAN UTAMA
# =====================================================
TEXT_COL = None
for candidate in ['clean_final', 'review', 'content', 'comment', 'text', 'body', 'reviewBody']:
    if candidate in df_real.columns:
        TEXT_COL = candidate
        break

if TEXT_COL is None:
    raise ValueError("‚ùå Kolom teks ulasan tidak ditemukan. Tentukan kolom teks utama secara manual.")

print(f"‚úÖ Kolom ulasan terdeteksi: '{TEXT_COL}'")

# Hapus data kosong atau hanya berisi spasi
df_real = df_real.dropna(subset=[TEXT_COL])
df_real = df_real[df_real[TEXT_COL].astype(str).str.strip() != ""]

# =====================================================
# 3. IDENTIFIKASI KOLOM UNTUK STRATIFIKASI (JIKA ADA)
# =====================================================
possible_strata = [c for c in ['lang', 'sentiment'] if c in df_real.columns]
if len(possible_strata) > 0:
    print(f"üîπ Kolom stratifikasi digunakan: {possible_strata}")
else:
    print("‚ÑπÔ∏è Tidak ada kolom 'lang' atau 'sentiment'. Sampling dilakukan secara acak.")

# =====================================================
# 4. FUNGSI PEMBUAT DATA DUMMY
# =====================================================
def make_dummy(df, pct=0.05, stratify_cols=None, text_col='review', random_state=42):
    """
    Membuat subset data dummy berdasarkan persentase tertentu.
    Jika kolom stratifikasi tersedia, sampling dilakukan secara proporsional.
    Augmentasi ringan ditambahkan untuk menambah variasi teks.
    """
    random.seed(random_state)
    np.random.seed(random_state)

    n_target = max(1, int(len(df) * pct))
    print(f"\nüìä Membuat dummy {pct*100:.0f}% ({n_target} data)...")

    # Sampling proporsional jika kolom stratifikasi ada
    if stratify_cols and all(c in df.columns for c in stratify_cols):
        df_temp = df.copy()
        df_temp['__key__'] = df_temp[stratify_cols].astype(str).agg('_'.join, axis=1)
        df_sample = (
            df_temp.groupby('__key__', group_keys=False)
            .apply(lambda x: x.sample(frac=pct, replace=len(x)*pct > len(x), random_state=random_state))
            .reset_index(drop=True)
        )
        df_sample.drop(columns=['__key__'], inplace=True, errors='ignore')
    else:
        df_sample = df.sample(n=n_target, random_state=random_state)

    # Tambahkan kolom default bila belum ada
    if 'lang' not in df_sample.columns:
        df_sample['lang'] = 'unknown'
    if 'sentiment' not in df_sample.columns:
        df_sample['sentiment'] = 'unknown'

    # Augmentasi ringan teks (menambah kata acak 20% peluang)
    def augment_text(t):
        t = str(t)
        if random.random() < 0.2:
            t += " " + random.choice(["gacha", "pity", "banner", "turn-based", "update", "lag"])
        return t

    df_sample[text_col] = df_sample[text_col].apply(augment_text)

    print(f"‚úÖ Dummy {pct*100:.0f}% selesai dibuat: {len(df_sample)} baris.")
    return df_sample

# =====================================================
# 5. PEMBUATAN DAN PENYIMPANAN DATA DUMMY
# =====================================================
output_dir = "./"
sizes = [0.05, 0.10, 0.25]  # 5%, 10%, dan 25% dari total data
dummies = {}

for pct in sizes:
    dummy_df = make_dummy(
        df_real,
        pct=pct,
        stratify_cols=possible_strata,
        text_col=TEXT_COL
    )
    out_path = os.path.join(output_dir, f"honkai_dummy_{int(pct*100)}pct.csv")
    dummy_df.to_csv(out_path, index=False, encoding='utf-8')
    dummies[pct] = out_path
    print(f"üíæ File disimpan ke: {out_path}")

# =====================================================
# 6. RANGKUMAN
# =====================================================
print("\nüéØ Semua file dummy berhasil dibuat dan disimpan:")
for k, v in dummies.items():
    print(f"- {int(k*100)}% ‚Üí {v}")
```

```python
# ==============================================
# PREPROCESSING UNTUK DATA DUMMY HONKAI
# ==============================================

import os
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from typing import List

# ----------------------------------------------
# Persiapan Resource
# ----------------------------------------------
nltk.download('punkt')
nltk.download('stopwords')

# Folder dan daftar dummy
INPUT_DIR = "./"
OUTPUT_DIR = "./dummy_preprocessed"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# File dummy yang akan diproses (otomatis deteksi)
dummy_files = [
    f for f in os.listdir(INPUT_DIR)
    if f.startswith("honkai_dummy_") and f.endswith(".csv")
]

if not dummy_files:
    raise FileNotFoundError("‚ùå Tidak ditemukan file dummy (honkai_dummy_*.csv) di folder kerja!")

print(f"üìÇ Ditemukan {len(dummy_files)} file dummy untuk diproses:\n" + "\n".join(dummy_files))

# ----------------------------------------------
# Normalisasi Kata Tidak Baku
# ----------------------------------------------
normalization_dict = {
    "ga": "tidak", "gak": "tidak", "gaada": "tidak ada", "bgt": "banget", "aja": "saja",
    "jd": "jadi", 'yg': 'yang', 'tpi': 'tapi', 'bg': 'bang', 'sy': 'saya', 'gk': 'tidak',
    'dpt': 'dapat', 'bngt': 'banget', 'lg': 'lagi', 'gw': 'saya', 'gue': 'saya', 'lu': 'kamu',
    'udh': 'sudah', 'udah': 'sudah', 'kalo': 'kalau', 'kyk': 'seperti', 'jg': 'juga', 
    'blm': 'belum', 'gmn': 'bagaimana', 'knp': 'kenapa', 'trs': 'terus', 'sm': 'sama', 
    'cmn': 'cuma', 'blum': 'belum', 'dapet': 'dapat', 'gem': 'game', 'relic': 'relic',
    'bnyk': 'banyak', 'mantep': 'mantap', 'muas': 'puas', 'pls': 'please', 'ny': 'nya',
    'pw': 'power', 'geme': 'game', 'gim': 'game'
}

# ----------------------------------------------
# Stopword Gabungan
# ----------------------------------------------
stop_words_id = set(stopwords.words('indonesian'))
stop_words_en = set(stopwords.words('english'))
additional_stopwords = {
    "game", "nya", "sih", "aja", "yang", "gw", "gue", "kak", "min", "makasih",
    "terimakasih", "keren", "ok", "tolong", "mohon", "banget", "si", "bagus", "keren",
    "mantap", "gamenya", "okelah", "thanks", "good", "nice", "best", "seru", "oke",
    "terima", "kasih", "semangat", "moga", "thank", "you", "please", "main", "baik",
    "ya", "sehat", "hoyoverse", "coy", "love", "ni", "suck", "like", "make", "play"
}
stop_words = stop_words_id.union(stop_words_en).union(additional_stopwords)

# ----------------------------------------------
# Setup Stemmer
# ----------------------------------------------
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# ----------------------------------------------
# Fungsi Preprocessing
# ----------------------------------------------
def case_folding(text: str) -> str:
    return str(text).lower()

def clean_text(text: str) -> str:
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def normalize_text(text: str) -> str:
    words = text.split()
    words = [normalization_dict.get(w, w) for w in words]
    return " ".join(words)

def tokenize_text(text: str) -> List[str]:
    return re.findall(r'\b[^\d\W_]+\b', text)

def remove_stopwords(tokens: List[str]) -> List[str]:
    return [t for t in tokens if t not in stop_words]

def stem_tokens(tokens: List[str]) -> List[str]:
    return [stemmer.stem(t) for t in tokens]

def preprocess_dataframe(df: pd.DataFrame, text_col: str = "review") -> pd.DataFrame:
    df = df.dropna(subset=[text_col])
    df = df[df[text_col].astype(str).str.strip() != ""]
    
    df["case_folding"] = df[text_col].apply(case_folding)
    df["cleaned"] = df["case_folding"].apply(clean_text)
    df["normalized"] = df["cleaned"].apply(normalize_text)
    df["token"] = df["normalized"].apply(tokenize_text)
    df["no_stopword"] = df["token"].apply(remove_stopwords)
    df["stemmed"] = df["no_stopword"].apply(stem_tokens)
    df["clean_final"] = df["stemmed"].apply(lambda x: " ".join(x))
    
    df = df[df["clean_final"].str.strip() != ""]
    return df

# ----------------------------------------------
# Jalankan Preprocessing untuk Tiap Dummy
# ----------------------------------------------
for file in dummy_files:
    file_path = os.path.join(INPUT_DIR, file)
    df = pd.read_csv(file_path)
    
    # Deteksi kolom teks
    text_col = next((c for c in ['review', 'content', 'comment', 'text', 'body', 'reviewBody'] if c in df.columns), None)
    if text_col is None:
        raise ValueError(f"‚ùå Kolom teks tidak ditemukan pada {file}")
    
    print(f"\nüîπ Memproses file: {file} (kolom teks: {text_col})")
    df_preprocessed = preprocess_dataframe(df, text_col=text_col)
    print(f"‚úÖ {len(df_preprocessed)} baris setelah preprocessing.")

    # Ekstrak ukuran dari nama file (misal 5 dari honkai_dummy_5pct.csv)
    pct_match = re.search(r"_(\d+)pct", file)
    pct_str = pct_match.group(1) if pct_match else "unknown"

    output_file = os.path.join(OUTPUT_DIR, f"dummy_preprocessed_{pct_str}pct.csv")
    df_preprocessed.to_csv(output_file, index=False, encoding="utf-8-sig")
    print(f"üìÅ Hasil disimpan: {output_file}")

print("\nüéØ Semua dummy telah diproses dan disimpan di folder './dummy_preprocessed/'")
```

```python
# ==============================================================
# TAHAP 2 & 3: TRANSFORMASI & PEMODELAN TOPIK (DATA DUMMY)
# ==============================================================
# Menerapkan IndoBERT + BERTopic untuk semua file dummy_preprocessed
# ==============================================================
import pandas as pd
import numpy as np
import os
import time
from sentence_transformers import SentenceTransformer
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
from umap import UMAP
from hdbscan import HDBSCAN
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re

# --------------------------------------------------------------
# 1Ô∏è LOAD SEMUA FILE DUMMY YANG SUDAH DIPREPROCESSING
# --------------------------------------------------------------
INPUT_DIR = "./dummy_preprocessed"
OUTPUT_ROOT = "./Hasil_dummy"
os.makedirs(OUTPUT_ROOT, exist_ok=True)

dummy_files = [
    f for f in os.listdir(INPUT_DIR)
    if f.startswith("dummy_preprocessed_") and f.endswith(".csv")
]

if not dummy_files:
    raise FileNotFoundError("‚ùå Tidak ditemukan file dummy_preprocessed_*.csv di folder dummy_preprocessed/")

print(f"üìÅ Ditemukan {len(dummy_files)} file dummy untuk diproses:\n" + "\n".join(dummy_files))

# --------------------------------------------------------------
# 2Ô∏è MEMUAT MODEL EMBEDDING (INDOBERT)
# --------------------------------------------------------------
local_model_path = "./model/indobert-base-p1"
if os.path.exists(local_model_path):
    print(f"\nüîπ Memuat model IndoBERT lokal dari '{local_model_path}'...")
    embedding_model = SentenceTransformer(local_model_path)
else:
    print("\n‚ö†Ô∏è Model lokal tidak ditemukan. Mengunduh dari Hugging Face...")
    embedding_model = SentenceTransformer("indobenchmark/indobert-base-p1")

# --------------------------------------------------------------
# 3Ô∏è FUNGSI PENENTU PARAMETER OTOMATIS BERDASARKAN JUMLAH DATA
# --------------------------------------------------------------
def auto_params(num_docs: int):
    if num_docs < 1000:
        return {"min_df": 1, "n_neighbors": 8, "min_cluster_size": 8, "min_samples": 5}
    elif num_docs < 3000:
        return {"min_df": 2, "n_neighbors": 10, "min_cluster_size": 15, "min_samples": 7}
    elif num_docs < 7000:
        return {"min_df": 3, "n_neighbors": 15, "min_cluster_size": 30, "min_samples": 10}
    elif num_docs < 15000:
        return {"min_df": 5, "n_neighbors": 20, "min_cluster_size": 50, "min_samples": 12}
    else:
        return {"min_df": 0.002, "n_neighbors": 20, "min_cluster_size": 60, "min_samples": 12}

# --------------------------------------------------------------
# 4Ô∏è LOOP UNTUK MEMPROSES SETIAP FILE DUMMY
# --------------------------------------------------------------
for dummy_file in dummy_files:
    pct_match = re.search(r"_(\d+)pct", dummy_file)
    pct_str = pct_match.group(1) if pct_match else "unknown"
    print(f"\n{'='*80}\nüîπ MEMPROSES DUMMY {pct_str}% ‚Üí {dummy_file}\n{'='*80}")

    # ------------------------------------------
    # LOAD DATA
    # ------------------------------------------
    df = pd.read_csv(os.path.join(INPUT_DIR, dummy_file))
    if "clean_final" not in df.columns:
        raise ValueError(f"Kolom 'clean_final' tidak ditemukan pada {dummy_file}. Pastikan hasil preprocessing benar.")
    docs = df["clean_final"].astype(str).tolist()
    print(f"Jumlah dokumen: {len(docs)}")

    # ------------------------------------------
    # EMBEDDINGS (CACHE PER FILE)
    # ------------------------------------------
    embedding_path = os.path.join(INPUT_DIR, f"embeddings_dummy_{pct_str}pct.npy")

    if os.path.exists(embedding_path):
        print("üì¶ Memuat embeddings dari cache...")
        embeddings = np.load(embedding_path)
    else:
        print("üß† Menghitung embeddings baru dengan IndoBERT...")
        start_embed = time.time()
        embeddings = embedding_model.encode(docs, batch_size=32, show_progress_bar=True)
        np.save(embedding_path, embeddings)
        print(f"üíæ Embeddings disimpan ke: {embedding_path}")
        print(f"‚è±Ô∏è Waktu komputasi: {(time.time()-start_embed)/60:.2f} menit")

    print(f"‚úÖ Shape embeddings: {embeddings.shape}")

    # ------------------------------------------
    # PARAMETER OTOMATIS
    # ------------------------------------------
    params = auto_params(len(df))
    print(f"üîß Parameter otomatis: {params}")

    # ------------------------------------------
    # KONFIGURASI BERTopic
    # ------------------------------------------
    vectorizer_model = CountVectorizer(max_df=0.8, min_df=params["min_df"], ngram_range=(1,3))
    umap_model = UMAP(
        n_neighbors=params["n_neighbors"], 
        n_components=5, 
        min_dist=0.0, 
        metric="cosine", 
        random_state=42
    )
    hdbscan_model = HDBSCAN(
        min_cluster_size=params["min_cluster_size"],
        min_samples=params["min_samples"],
        metric="euclidean",
        cluster_selection_method="eom",
        prediction_data=True
    )

    topic_model = BERTopic(
        embedding_model=embedding_model,
        umap_model=umap_model,
        hdbscan_model=hdbscan_model,
        vectorizer_model=vectorizer_model,
        language="multilingual",
        calculate_probabilities=True,
        verbose=True
    )

    # ------------------------------------------
    # FIT MODEL
    # ------------------------------------------
    print("üöÄ Melatih model BERTopic...")
    start_time = time.time()
    topics, probs = topic_model.fit_transform(docs, embeddings=embeddings)
    duration = (time.time() - start_time) / 60
    print(f"‚úÖ Pelatihan selesai dalam {duration:.2f} menit.")

    # --------------------------------------------------------------
    # EVALUASI MODEL (KUALITATIF & KUANTITATIF)
    # --------------------------------------------------------------
    topic_info = topic_model.get_topic_info()
    print("Informasi Topik Utama:")
    print(topic_info.head())
    
    # Hitung koherensi topik
    topics_for_eval = []
    for topic_id in topic_info[topic_info["Topic"] != -1]["Topic"]:
        top_words = topic_model.get_topic(topic_id)
        if top_words and isinstance(top_words, list):
            # Pastikan elemen valid dan hanya ambil kata
            words = [w for w, _ in top_words if isinstance(w, str)]
            if len(words) > 1:  # minimal 2 kata agar bisa dihitung koherensinya
                topics_for_eval.append(words)
    
    if len(topics_for_eval) > 1:
        tokenized_docs = [doc.split() for doc in docs if isinstance(doc, str) and len(doc.split()) > 2]
        dictionary = Dictionary(tokenized_docs)
        corpus = [dictionary.doc2bow(toks) for toks in tokenized_docs]
        cm = CoherenceModel(
            topics=topics_for_eval,
            texts=tokenized_docs,
            dictionary=dictionary,
            coherence="c_v"
        )
        coherence_score = cm.get_coherence()
        print(f"üìà Skor Koherensi Rata-rata (c_v): {coherence_score:.4f}")
    else:
        print("‚ö†Ô∏è Tidak cukup topik yang valid untuk menghitung skor koherensi.")

    # --------------------------------------------------------------
    # 8Ô∏è VISUALISASI WORDCLOUD PER TOPIK
    # --------------------------------------------------------------
    print("\n Membuat visualisasi...")
    
    # --- A. Bar Chart Distribusi Topik ---
    fig_bar = topic_model.visualize_barchart(top_n_topics=10)
    fig_bar.show()
    
    # --- B. Intertopic Distance Map ---
    fig_map = topic_model.visualize_topics()
    fig_map.update_layout(
       xaxis={
            "visible": True,
            "showticklabels": True,
            "showgrid": True
        },
        yaxis={
            "visible": True,
            "showticklabels": True,
            "showgrid": True
        }
    )
    fig_map.show()
    
    # --- C. WordCloud untuk Setiap Topik (Semua disimpan, hanya 3 ditampilkan) ---
    output_wc_dir = os.path.join(output_dir, "wordclouds")
    os.makedirs(output_wc_dir, exist_ok=True)
    
    print("\nMembuat WordCloud untuk setiap topik (10 kata teratas)...")
    
    # Urutkan berdasarkan ukuran topik (jumlah dokumen)
    topic_info_sorted = topic_info[topic_info["Topic"] != -1].sort_values("Count", ascending=False)
    top_display_n = 3  # hanya tampilkan 3 di output
    
    for i, topic_id in enumerate(topic_info_sorted["Topic"]):
        # Ambil 10 kata teratas berdasarkan bobot
        top_words = topic_model.get_topic(topic_id)[:10]
        freq_dict = dict(top_words)
        
        # Buat WordCloud
        wordcloud = WordCloud(
            width=800, height=400,
            background_color="white",
            colormap="viridis"
        ).generate_from_frequencies(freq_dict)
        
        # Simpan ke file
        save_path = os.path.join(output_wc_dir, f"wordcloud_topic_{topic_id}.png")
        plt.figure(figsize=(10, 5))
        title = topic_info.loc[topic_info.Topic == topic_id, "Name"].iloc[0]
        plt.title(f"WordCloud Topik #{topic_id}: {title}")
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        plt.savefig(save_path, bbox_inches='tight')
        plt.close()  # ditutup agar tidak menumpuk memori
    
        # Tampilkan hanya 3 WordCloud teratas
        if i < top_display_n:
            plt.figure(figsize=(10, 5))
            plt.title(f"Top {i+1}: WordCloud Topik #{topic_id}: {title}")
            plt.imshow(wordcloud, interpolation="bilinear")
            plt.axis("off")
            plt.show()
    
    print(f"\n‚úÖ Semua WordCloud ({len(topic_info_sorted)} topik) disimpan di folder: {output_wc_dir}")
    print(f"üìä Hanya {top_display_n} WordCloud teratas yang ditampilkan di output.")


    # ------------------------------------------
    # SIMPAN HASIL
    # ------------------------------------------
    topic_info.to_csv(os.path.join(output_dir, f"hasil_analisis_topik_{pct_str}pct.csv"), index=False, encoding="utf-8-sig")

    rep_docs = []
    for topic_id in topic_info[topic_info["Topic"] != -1]["Topic"]:
        docs_rep = topic_model.get_representative_docs(topic_id)
        for d in docs_rep:
            rep_docs.append({"Topic_ID": topic_id, "Representative_Doc": d})
    pd.DataFrame(rep_docs).to_excel(os.path.join(output_dir, f"dokumen_representatif_{pct_str}pct.xlsx"), index=False)

    summary_data = {
        "Persentase_Data": f"{pct_str}%",
        "Jumlah_Dokumen": len(docs),
        "Jumlah_Topik": unique_topics,
        "Rasio_Outlier": outlier_ratio,
        "Koherensi_c_v": coherence_score
    }
    pd.DataFrame([summary_data]).to_csv(os.path.join(output_dir, f"ringkasan_{pct_str}pct.csv"), index=False)

    print(f"üìä Ringkasan hasil disimpan di: {output_dir}")

print("\nüéØ Semua analisis dummy selesai dan disimpan di folder './Hasil_dummy/'")
```

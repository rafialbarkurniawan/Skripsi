#### Kode Scraping dengan batch

```python
import pandas as pd
from google_play_scraper import reviews, Sort
import time

def scrape_reviews_batch(app_id, batch_size=25000, total_limit=50000, sleep_time=1, sort=Sort.NEWEST):
    """
    Scrape ulasan Google Play Store dalam batch agar aman jika terhenti.
    
    Parameters:
    - app_id (str): ID aplikasi di Google Play (contoh: 'com.miHoYo.hkrpgoversea')
    - batch_size (int): Jumlah ulasan per batch
    - total_limit (int): Total ulasan yang ingin diambil
    - sleep_time (int): Jeda antar batch (detik) untuk mencegah block
    - sort (Sort): Urutan ulasan (Sort.NEWEST atau Sort.MOST_RELEVANT)
    
    Output:
    - File CSV per batch: reviews_batch_1.csv, reviews_batch_2.csv, dst.
    """
    
    all_reviews = []
    count = 0
    batch_num = 1
    
    while count < total_limit:
        # Ambil batch ulasan
        print(f"ðŸ”„ Mengambil batch {batch_num} ...")
        
        rvs, _ = reviews(
            app_id,
            lang='id',       # Bahasa ulasan
            country='id',    # Server Indonesia
            sort=sort,
            count=batch_size,
            filter_score_with=None
        )
        
        # Simpan ke dataframe
        df_batch = pd.DataFrame(rvs)
        filename = f"reviews_batch_{batch_num}.csv"
        df_batch.to_csv(filename, index=False, encoding='utf-8-sig')
        
        print(f"âœ… Batch {batch_num} selesai. Disimpan di {filename} ({len(df_batch)} ulasan).")
        
        # Tambahkan ke total
        all_reviews.extend(rvs)
        count += len(rvs)
        
        # Jeda sebentar agar tidak dianggap spam
        time.sleep(sleep_time)
        
        batch_num += 1
        
        # Jika data kurang dari batch_size â†’ artinya tidak ada ulasan lagi
        if len(rvs) < batch_size:
            print("ðŸš¨ Tidak ada ulasan tambahan. Scraping selesai.")
            break
    
    # Simpan semua gabungan
    df_all = pd.DataFrame(all_reviews)
    df_all.to_csv("reviews_all.csv", index=False, encoding='utf-8-sig')
    print(f"ðŸ“ Semua data ulasan tersimpan di reviews_all.csv ({len(df_all)} ulasan).")

# ==============================
# ðŸ“Œ CONTOH PENGGUNAAN
# ==============================

# Honkai Star Rail
app_id = "com.miHoYo.hkrpgoversea"

# Magic Chess Go Go
# app_id = "com.mobile.legends.chess"  # â† contoh, pastikan ID aplikasinya benar

scrape_reviews_batch(app_id, batch_size=25000, total_limit=50000, sort=Sort.NEWEST)
```
